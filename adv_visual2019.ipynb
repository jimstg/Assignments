```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('states_all.csv')
df.head()
```

```python
df['STATE_REVENUE']= df['STATE_REVENUE'].fillna(0)
df['TOTAL_REVENUE']= df['TOTAL_REVENUE'].fillna(0)
df['LOCAL_REVENUE']= df['LOCAL_REVENUE'].fillna(0)
df['FEDERAL_REVENUE']= df['FEDERAL_REVENUE'].fillna(0)
```

```python
washst = df.loc[(df['STATE']=='WASHINGTON')]
washstrev = df.loc[(df['STATE']=='WASHINGTON'),'STATE_REVENUE']
washsttot = df.loc[(df['STATE']=='WASHINGTON'),'TOTAL_REVENUE']
washstloc = df.loc[(df['STATE']=='WASHINGTON'),'LOCAL_REVENUE']
washstfed = df.loc[(df['STATE']=='WASHINGTON'),'FEDERAL_REVENUE']
orestfed = df.loc[(df['STATE']=='OREGON'),'FEDERAL_REVENUE']
orestloc = df.loc[(df['STATE']=='OREGON'),'LOCAL_REVENUE']
orestrev = df.loc[(df['STATE']=='OREGON'),'STATE_REVENUE']
oresttot = df.loc[(df['STATE']=='OREGON'),'TOTAL_REVENUE']
washst
```

```python
plt.hist(washstrev, color='red',  alpha=.5,label='Wash State Rev')
plt.hist(orestrev, color='blue', alpha=.5, label='Oregon State Rev')
plt.show()
```

```python
sns.set(style="whitegrid")

ax = sns.boxplot(x= "YEAR", y= "STATE_REVENUE", data=washst, palette='pastel')

ax.set(xlabel='Washington')
plt.show()
```

```python
n = sns.lmplot(y = 'YEAR',
               x = 'TOTAL_REVENUE',
              data = washst,
              fit_reg=True,
              scatter_kws={'alpha':0.4})
plt.show()
```

```python
n = sns.factorplot(x = 'YEAR', y = 'STATE_REVENUE', data=washst, size=6, kind="point", palette='pastel', ci=95, join=False)
n.despine(left=True)
n.set_ylabels("State Revenue")
n.set_xlabels("Year")
plt.title("Pointplot: State Revenue Per Year")
plt.show()
```

```python
n = sns.factorplot(x = 'YEAR', y = 'STATE_REVENUE', data=washst, size=6, kind="bar", palette='pastel')
n.despine(left=True)
n.set_ylabels("State Revenue")
n.set_xlabels("Year")
plt.title("Barplot: State Revenue Per Year")
plt.show()
```
